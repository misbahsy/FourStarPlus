```python
# For multiple pages

data_list = []
url_dict = {'Cellphones And Accesories': cell_acc,'Televisions': tv, 'Cameras': camera, 'Computer Accesories': computer_acc, 'Computers And Tablets': computers_tablets }
for url_key, url_value in url_dict.items():
    print(url_key)
    for page in range(1,10):
        print(page)

        options = Options()
        options.set_headless(headless=True)

        driver = webdriver.Chrome('/Users/Misbah/Learning/byteacademy/chromedriver', options=options)  
        driver.get(url_value)

        data = driver.page_source
        driver.close()

        soup = BeautifulSoup(data, 'html.parser')

        data_divs = soup.find_all('div', {'class': 's-result-item s-asin sg-col-0-of-12 sg-col-16-of-20 sg-col sg-col-12-of-16'})
        len(data_divs)


        for data_div in data_divs:
            data_dict = {}
            data_dict['Category'] = url_key
            try:
                data_dict['Product'] = data_div.find('span', {'class': 'a-size-medium a-color-base a-text-normal'}).text
            except:
                data_dict['Product'] = None
            try:
                data_dict['Rating'] = data_div.find('span', {'class': 'a-icon-alt'}).text
            except:
                data_dict['Rating'] = None
            try:
                data_dict['Reviews'] = data_div.find('span', {'class': 'a-size-base'}).text
            except:
                data_dict['Reviews'] = None
            try:
                data_dict['Price'] = data_div.find('span', {'class': 'a-price'}).text.split('$')[1]
            except:
                data_dict['Price'] = None
            try:
                data_dict['Image'] = data_div.find('img')['src']
            except:
                data_dict['Image'] = None
            try:
                data_dict['ASIN'] = data_div['data-asin']
            except:
                data_dict['ASIN'] = None
            try:
                data_dict['Link'] = 'https://www.amazon.com'+ data_div.find('a', {'class': 'a-link-normal a-text-normal'})['href']
            except:
                data_dict['Link'] = None
            data_list.append(data_dict)

data_list

df_full = pd.DataFrame(data_list)
df_full
```

```python
df_temp = df_full.copy()
df_temp['Reviews'] = df_temp['Reviews'].str.replace(',', '')
df_temp['Reviews'] = df_temp['Reviews'].astype('int')
df_sorted = df_temp.sort_values(by=['Reviews'], ascending=False)
df_sorted.head(30)
```

```python
df_sorted['Rating_Int'] = df_sorted['Rating'].str.split(' ', expand=True)[0]
df_sorted[['Rating_Int']] = df_sorted[['Rating_Int']].astype('float32')
df_sorted
```

```python
df_sorted['Price'] = df_sorted['Price'].str.replace(',', '')
```

```python
df_sorted['Price'] = df_sorted['Price'].astype('float32')
```
